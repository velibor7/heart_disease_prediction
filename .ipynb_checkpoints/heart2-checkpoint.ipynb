```python
def predict(row, coefficients):
    yhat = coefficients[0]
    for i in range(len(row) -1):
        yhat += coefficients[i+1] * row[i]
    return 1.0 / (1.0 + np.exp(-yhat))
```

```python
# Estimate logistic regression coefficients using stochastic gradient descent
def coefficients_sgd(train, l_rate, n_epoch):
    coef = [0.0 for i in range(len(train[0]))]
    for epoch in range(n_epoch):
        sum_error = 0
        for row in train:
            yhat = predict(row, coef)
            error = row[-1] - yhat
            sum_error += error**2
            coef[0] = coef[0] + l_rate * error * yhat * (1.0 - yhat)
            for i in range(len(row)-1):
                coef[i + 1] = coef[i + 1] + l_rate * error * yhat * (1.0 - yhat) * row[i]
        print('>epoch=%d, lrate=%.3f, error=%.3f' % (epoch, l_rate, sum_error))
    return coef
```

```python
# FUNCIONS TO NORMALIZE THE DATASET
# trazimo minimum i maximum svake kolone, da bi mogli da normalizujemo podatke
def data_minmax(dataset):
    minmax = list()
    for i in range(len(dataset[0])):
        col_values = [row[i] for row in dataset]
        value_min = min(col_values)
        value_max = max(col_values)
        minmax.append([value_min, value_max])
    return minmax

def normalize_dataset(dataset, minmax):
    for row in dataset:
        for i in range(len(row)):
            row[i] = (row[i] - minmax[i][0]) / (minmax[i][1] - minmax[i][0])
```

```python
# split dataset into k folds
def k_folds_cross_validation_split(dataset, n_folds):
    dataset_split = list()
    dataset_copy = list(dataset) # odavde izvlacimo random redove
    fold_size = int(len(dataset) / n_folds) # broj redova / k
    for i in range(n_folds):
        fold = list()
        while len(fold) < fold_size:
            index = randrange(len(dataset_copy))
            fold.append(dataset_copy.pop(index))
        dataset_split.append(fold)
    return dataset_split # dobijamo podeljenu datu u n_folds delova
```

```python
# calculating the accuracy percentage
def accuracy_metrics(actual, predicted):
    correct = 0
    for i in range(len(actual)):
        if actual[i] == predicted[i]:
            correct += 1
    return correct / float(len(actual)) * 100.0
```

```python
# evaluating the algorithm
def evaluate(dataset, algorithm, n_folds, *args):
    folds = k_folds_cross_validation_split(dataset, n_folds) # lista foldova
    scores = list() # ovde cuvamo acc od svakog folda posebno
    for fold in folds: # fold je lista redova, deo podataka
        train_set = list(folds)
        train_set.remove(fold) # izbacicemo 1, da bi radili sa k - 1
        train_set = sum(train_set, [])
        test_set = list()
        for row in fold: # za svaki red u tom chunku(foldu) date
            row_copy = list(row) # napravimo kopiju rowa sa kojom radimo
            test_set.append(row_copy)
            row_copy[-1] = None # poslednji value nam je target
        predicted = algorithm(train_set, test_set, *args) # predictujemo target tog rowa
        actual = [row[-1] for row in fold]
        accuracy = accuracy_metrics(actual, predicted) # racunamo acc
        scores.append(accuracy) # dodajemo acc za taj row u scores
        # print("Fold no." + fold + ". Acc: " + accuracy);
    return scores # vraca acc za svaki fold posebno
```
