```python
import csv
import pandas as pd
import numpy as np
```

```python
cleveland = pd.read_csv("./cleveland.csv")
framingham = pd.read_csv("./framingham.csv")
```

```python
print(cleveland.shape)
print(framingham.shape)
```

```python
cleveland.head()
cleveland.columns
# objasnjenja svih kolona za cleveland, izvuceno sa mesta gde je skinut originalan dataset
# gledamo gde sve mozemo da spojimo 2 dataset
# *moze age
# *moze 4 sex: sex (1 = male; 0 = female) *moze da se spoji
# 9 cp: chest pain type -- Value 1: typical angina -- Value 2: atypical angina -- Value 3: non-anginal pain -- Value 4: asymptomatic 
# 10 trestbps: resting blood pressure (in mm Hg on admission to the hospital) 
# *moze  12 chol: serum cholestoral in mg/dl 
# 15 years (number of years as a smoker) 
# 16 fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) 
# 19 restecg: resting electrocardiographic results -- Value 0: normal -- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV) -- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria 
# 32 thalach: maximum heart rate achieved
# 38 exang: exercise induced angina (1 = yes; 0 = no)
# 40 oldpeak = ST depression induced by exercise relative to rest 
# 41 slope: the slope of the peak exercise ST segment -- Value 1: upsloping -- Value 2: flat -- Value 3: downsloping ldv5: height at rest 
# 44 ca: number of major vessels (0-3) colored by flourosopy
# 51 thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
# *moze 58 num: diagnosis of heart disease (angiographic disease status) -- Value 0: < 50% diameter narrowing -- Value 1: > 50% diameter narrowing *moze da se spoji
```

```python
framingham.head()
framingham.columns
# *moze age, 
# *moze male(0 – zensko, 1 - musko), 
# education (po meni nebitno), 
# currentSmoker (0 – nepusac, 1 - pusac), 
# cigsPerDay, BPMeds (lekovi za visok pritisak, 0 – ne uzima, 1 - uzima), 
# prevalentStroke (da li je osoba do sad imala srcani udar 0/1), 
# prevalentHyp (da li je osoba do sad imala visoki krvni pritisak 0/1), 
# diabetes (0 – nema, 1 - ima), totChol (nivo holesterola u krvi mg/dL), 
# sysBP(systolic? mmHg), 
# diaBP(diastolic? mmHg), 
# BMI (“Body Mass Index” = tezina/visina), 
# heartRate (brzina kucanja srca beats/min), 
# glucose (mg/dL), 
# * mozeTenYearCHD (target)
```

```python
# data will be these 2 datasets merged
# prilagodjavamo imena kolona fiminghamovog dataseta, da bismo bolje spojili datasetove
# u friminghamu male prestavlja pol, u obe tabele 1 - male, 0 - woman
# TenYearCHD je target value, u obe tabele 1 prestavlja da su dobili/imaju bolest srce
framingham = framingham.rename(columns={'male': 'sex', 'TenYearCHD': 'target'})
framingham.head()
data = pd.merge(cleveland, framingham)
```

```python
# stavljamo target na poslednje mesto cisto da nam bude lakse
data = data.reindex(columns=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'education',
       'currentSmoker', 'cigsPerDay', 'BPMeds', 'prevalentStroke',
       'prevalentHyp', 'diabetes', 'totChol', 'sysBP', 'diaBP', 'BMI',
       'heartRate', 'glucose', 'target'])
data.head()
```

```python
data = data.dropna()
data.isna().sum()
```

```python
data.columns
```

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
X = data.iloc[:, 0:26]
y = data.iloc[:, -1]
test = SelectKBest(score_func=chi2, k=15)
fit = test.fit(X, y)
data_scores = pd.DataFrame(fit.scores_)
data_columns = pd.DataFrame(X.columns)

final_scores = pd.concat([data_columns, data_scores], axis=1)
final_scores.columns = ['Specs', 'Score']
```

```python
final_scores = final_scores.sort_values(by='Score', ascending=False)
final_scores
```

```python
# uzecemo prvih 15
# ne znam da li je problem zato sto ne sadrze ni age ni sex?
features_list = final_scores["Specs"].tolist()[:15]
features_list
```

```python
data = data[features_list]
data.head()
```
